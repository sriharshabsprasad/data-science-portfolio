```python
# Load the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from prophet import Prophet

import warnings
warnings.filterwarnings("ignore") 
```

# Data Acquisition and Loading

```python
# read a csv file and create a dataframe
file_path = "data/us_retail_sales.csv"  # Update the path if necessary
retail_data = pd.read_csv(file_path)
```

```python
# Display the top 10 rows of a dataframe
print("First ten rows of the dataset:")
retail_data.head(10)
```

```python
print("Dimensions of the data frame:", retail_data.shape)
```

```python
# Reshape the data into a time series format
retail_data = retail_data.set_index("YEAR").stack().reset_index()
retail_data.columns = ["Year", "Month", "Sales"]
retail_data["Date"] = pd.to_datetime(retail_data["Year"].astype(str) + "-" + retail_data["Month"], format="%Y-%b")
retail_data = retail_data[["Date", "Sales"]]
```

```python
retail_data
```

# 1. Plot the data with proper labeling and make some observations on the graph.

```python
# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(retail_data["Date"], retail_data["Sales"], label="Retail Sales")
plt.title("Monthly Retail Sales in the US (1992 - 2021)")
plt.xlabel("Year")
plt.ylabel("Sales (in millions)")
plt.legend()
plt.grid(True)
plt.show()
```

## Key observations from the graph:
- From 1992 to 2021, retail sales in the US have generally trended upward.
- There are some noticeable fluctuations, particularly a sharp dip around the 2008 financial crisis and the COVID-19 pandemic in 2020.
- Following the COVID-19 downturn, retail sales saw a strong rebound, indicating a phase of recovery.

# 2. Split this data into a training and test set. Use the last year of data (July 2020 – June 2021) of data as your test set and the rest as your training set.

```python
# Prepare the training and test sets
retail_data["Timestamp"] = (retail_data["Date"] - retail_data["Date"].min()).dt.days  # Convert date to numeric
train = retail_data[retail_data["Date"] < "2020-07-01"]
test = retail_data[retail_data["Date"] >= "2020-07-01"]

X_train = train[["Timestamp"]]
y_train = train["Sales"]
X_test = test[["Timestamp"]]
y_test = test["Sales"]
```

# 3. Use the training set to build a predictive model for the monthly retail sales.

# Model 1 - Linear Regression

```python
# Build and train the model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

# 4. Use the model to predict the monthly retail sales on the last year of data.

```python
# Predict on the test set
y_pred = linear_model.predict(X_test)
```

```python
y_pred
```

# 5. Report the RMSE of the model predictions on the test set.

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Test set - Linear Regression Predictions - RMSE: {rmse:.2f}")
```

```python
# Plot the predictions
plt.figure(figsize=(12, 6))
plt.plot(train["Date"], y_train, label="Training Data", color="blue")
plt.plot(test["Date"], y_test, label="Test Data", color="green")
plt.plot(test["Date"], y_pred, label="Predictions", color="red", linestyle="--")
plt.title("Retail Sales Predictions with Linear Regression")
plt.xlabel("Year")
plt.ylabel("Sales (in millions)")
plt.legend()
plt.grid(True)
plt.show()
```

# Model 2 - Facebook Prophet

### Prophet is specifically designed to handle trends and seasonality, making it ideal for retail sales data.

### Key Features of the Prophet Model:
- Seasonality: Automatically captures yearly patterns.
- Trends: Handles long-term growth trends and holidays well.
- Flexibility: Robust against missing data and outliers.

```python
retail_data = retail_data[["Date", "Sales"]]

# Prepare the data for Prophet
retail_data_prophet = retail_data.rename(columns={"Date": "ds", "Sales": "y"})

# Split the data into training and testing sets
train = retail_data_prophet[retail_data_prophet["ds"] < "2020-07-01"]
test = retail_data_prophet[retail_data_prophet["ds"] >= "2020-07-01"]

# Build and fit the Prophet model
prophet_model = Prophet()
prophet_model.fit(train)
```

```python
# Create a dataframe for predictions
future = prophet_model.make_future_dataframe(periods=len(test)+1, freq="M")
forecast = prophet_model.predict(future)
```

```python
# Extract predictions for the test set
forecast_test = forecast[forecast["ds"] >= "2020-07-01"]
y_pred = forecast_test["yhat"].values
y_test = test["y"].values
```

```python
y_pred
```

```python
# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"Test set - Facebook Prophet Predictions - RMSE: {rmse:.2f}")
```

```python
# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train["ds"], train["y"], label="Training Data", color="blue")
plt.plot(test["ds"], test["y"], label="Test Data", color="green")
plt.plot(forecast_test["ds"], forecast_test["yhat"], label="Predictions", color="orange", linestyle="--")
plt.title("Retail Sales Forecast with Facebook Prophet")
plt.xlabel("Year")
plt.ylabel("Sales (in millions)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Prophet forecast components
prophet_model.plot_components(forecast)
plt.show()
```

- ### Test set - Linear Regression Predictions - RMSE: 66429.10
- ### Test set - Facebook Prophet Predictions - RMSE: 56029.51
- ### Prophet's predictions are more accurate than Linear Regression's, as indicated by a lower Root Mean Squared Error (RMSE).
